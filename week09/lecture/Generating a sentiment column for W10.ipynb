```python
from pathlib import Path
from textblob import TextBlob
csal_sentiments = []

folder_path = "csal/" # We're telling the code to look in the "csal/" subfolder, where the CSAL files all live.

for file_path in sorted(Path(folder_path).glob('*.txt')):
    
    text = open(file_path, encoding='utf-8').read()
    blob = TextBlob(text)
    csal_sentiments.append(blob.polarity)
```

```python
len(csal_sentiments)
```

```python
import pandas as pd

csal_meta_df = pd.read_csv('csal_meta_gender_signal.csv')
```

```python
csal_meta_df = csal_meta_df.sort_values(by='Text', ascending=True)
```

```python
csal_meta_df['polarity'] = csal_sentiments
```

```python
csal_meta_df.to_csv("csal_meta_gender_signal_polarity.csv", encoding='utf-8', index=False)
```

```python
csal_meta_df.sort_values(by='polarity', ascending=True)
```

```python
csal_meta_df.groupby('Author').agg({
    'polarity': 'mean'
}).sort_values(by='polarity')
```
